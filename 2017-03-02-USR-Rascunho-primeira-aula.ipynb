#Array es un vetor com elementos homogeneos: INT, FLOAT ...
Shape (filas y columasn) e type. 
Matriz sempre é de 2 dimensoes
Array pode ser de 1 a n dimensoes

```python
import numpy as np
a=np.array([1,2,3,4,5,6])
print(a[::2])
print(a.shape)
print(a.dtype)
```

```python
a=np.array([[1,2,3,4],
          [5,6,7,8]])
n,m=a.shape
print(a.shape, a.dtype)
print(a)
print(a[1,2])
print(a[0:n,0:m])
print(a[:,:])
#só ultima coluna mas continuando como matriz: 2 dimensoes
print('só a última coluna')
print(a[:,-1:])
print('só as 2 primeiras colunas')
print(a[:,0:2])

#ao reves
print('colunas e linhas invertidas')
print(a[::-1,::-1])
```

```python
import numpy as np
from numpy.random import rand
a=(rand(300,400)*255).astype(np.uint8)
print(a)
```
